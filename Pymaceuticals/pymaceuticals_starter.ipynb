## Observations and Insights 

1. Based on the data obtained, the number of mice for each drug regimen are kept the same. The gender distribution are also kept  the same.

2. It was found that mice that are treated with Capomulin ended up with lower tumor volume compare to Infubinol and Ceptamine treatment. It was also found that mice that are treated with Capomulin have a more consistent result (lower variance as smaller box and shorter whisker in the box plot) compare to Ramicane treatment

3. It was found that the heavier a mouse is, the larger the tumor volume.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df_mousestudy = mouse_metadata.merge(study_results, how='outer', on="Mouse ID")

# Display the data table for preview
df_mousestudy
```

```python
# Checking the number of mice.
num_of_mice = len(df_mousestudy["Mouse ID"].unique())
print(f"The number of mice is {num_of_mice}.")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
df_mouse_with_duplicates = df_mousestudy[df_mousestudy.duplicated(subset = ['Mouse ID', 'Timepoint']) == True]
mouse_deleted = df_mouse_with_duplicates["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
df_mouse_with_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

# Create a copy of original dataframe
df_mousestudy_cleaned = df_mousestudy.copy()

# Using for loop to traverse the mouse ID that can't be used because of duplicate data
for mouse in mouse_deleted:
    df_mousestudy_cleaned = df_mousestudy_cleaned.drop(df_mousestudy_cleaned[df_mousestudy_cleaned["Mouse ID"] == mouse].index)

# Preview of the cleaned data that will be used throughout
df_mousestudy_cleaned
```

```python
# Checking the number of mice in the clean DataFrame.
new_num_of_mice = len(df_mousestudy_cleaned["Mouse ID"].unique())

print(f"The number of mice being studied after cleaning the data set is {new_num_of_mice}.")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

df_mean_tumor = df_mousestudy_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby("Drug Regimen").mean()
df_median_tumor = df_mousestudy_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby("Drug Regimen").median()
df_var_tumor = df_mousestudy_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby("Drug Regimen").var()
df_std_tumor = df_mousestudy_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby("Drug Regimen").std()
df_sem_tumor = df_mousestudy_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby("Drug Regimen").sem()

df_summ = df_mean_tumor.copy()
df_summ = df_summ.rename(columns={"Tumor Volume (mm3)" : "Mean of Tumor Volume (mm3)"})
df_summ["Median of Tumor Volume (mm3)"] = df_median_tumor["Tumor Volume (mm3)"]
df_summ["Variance of Tumor Volume (mm3)"] = df_var_tumor["Tumor Volume (mm3)"]
df_summ["Standard Deviation of Tumor Volume (mm3)"] = df_std_tumor["Tumor Volume (mm3)"]
df_summ["SEM of Tumor Volume (mm3)"] = df_sem_tumor["Tumor Volume (mm3)"]


df_summ
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

# Create a copy of original dataframe
df_quarts = df_mousestudy_cleaned.copy()

# Drop all duplicate mouse ID and save the last (this can be done as the data is sorted by mouse ID with ascending Timepoint meaning last of the mouse ID is the last timepoint)
df_quarts = df_quarts.drop_duplicates("Mouse ID", keep = 'last')

df_quarts.head()

```

## Line and Scatter Plots

```python
# Generate a copy of data frame with only mouse that is treated with Capomulin
df_mouse_capo = df_mousestudy_cleaned.copy()
df_mouse_capo = df_mouse_capo[df_mouse_capo["Drug Regimen"] == "Capomulin"]

# Obtaining 1 mouse to plot the line graph randomly
random_mouse = df_mouse_capo["Mouse ID"].sample()

# Creating the dataframe for the one mouse treated with Capomulin
df_mouse_line = df_mouse_capo[df_mouse_capo["Mouse ID"] == random_mouse.values[0]]


# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
fig4, ax4 = plt.subplots(figsize = (8,8))
ax4.plot(df_mouse_line["Timepoint"], df_mouse_line["Tumor Volume (mm3)"], color='b', marker="o");
ax4.set(xlabel="Timepoint",
        ylabel="Tumor Volume (mm3)",
        title="Effect of Capomulin on a Mouse's Tumor");
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
df_mouse_scatter = df_mouse_capo[["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]]

df_mouse_scatter = df_mouse_scatter.groupby("Mouse ID").mean()
df_mouse_scatter.head()

fig5, ax5 = plt.subplots(figsize=(8,8))
ax5.scatter(df_mouse_scatter["Weight (g)"], df_mouse_scatter["Tumor Volume (mm3)"])
ax5.set(xlabel = "Weight of Mouse (g)", ylabel = "Average Tumor Volume (mm3)", title = "Correlation of Weight vs Tumor Volume");
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = df_mouse_scatter["Weight (g)"]
y_values = df_mouse_scatter["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f'''
The equation of the linear regression model is: {line_eq}
The correlation coefficient is: r={rvalue}
''')
```
